# Building a model to classify an image of Cat and Dog.

## 1.Importing the libraries

```python
import os
import numpy as np
import random
import itertools
import tensorflow as tf

config = tf.compat.v1.ConfigProto()
config.gpu_options.allow_growth = True
sess = tf.compat.v1.Session(config=config)

import shutil
from keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from keras.models import Sequential
from keras.layers import Conv2D,MaxPool2D,Flatten,Dense
from keras.optimizers import Adam
from keras.losses import categorical_crossentropy
from keras.callbacks import ReduceLROnPlateau
from keras.models import load_model
from matplotlib.pyplot import imshow
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix,classification_report
```

## 2.Creating directories for appropriate labels.

```python
#Specifing the classifications labels.
labels = ['cat','dog']
labels.sort() #To keep labels in uniformity

#Assigning the source path containg the label data(folders of images of cats and dogs)
data = 'dogs-v-cats/dataset/'

#Specifing the amount of train and test data.
split_size =0.98 # 90%train data, 10%test data 
```

```python
#Making directories to store the test and train data for model.
try:
    for i in labels:
        os.makedirs('labels/train/{}/'.format(i))
        os.makedirs('labels/val/{}/'.format(i))
        os.makedirs('labels/test/{}/'.format(i))
except OSError:
    pass
```

```python
for label in labels:
    img_lis = os.listdir(data+label)
    for i in range(len(img_lis)):
        if i < len(img_lis) * split_size:
            shutil.copyfile('dogs-v-cats/dataset/{}/{}'.format(label,img_lis[i]) , 'labels/train/{}/{}'.format(label,img_lis[i]))
        elif i < len(img_lis) * (split_size+0.01):
            shutil.copyfile('dogs-v-cats/dataset/{}/{}'.format(label,img_lis[i]) , 'labels/val/{}/{}'.format(label,img_lis[i]))
        else:
            shutil.copyfile('dogs-v-cats/dataset/{}/{}'.format(label,img_lis[i]) , 'labels/test/{}/{}'.format(label,img_lis[i]))
```

## 3.Using ImageGenerator for Image Augmentation and to load the data with appropriate labels. 

```python
train_datagen = ImageDataGenerator(rescale =1.0/255,
                                  rotation_range =40,
                                  width_shift_range =0.2,
                                  height_shift_range = 0.2,
                                  shear_range =0.2,
                                  zoom_range = 0.2,
                                  horizontal_flip =True,
                                  fill_mode = 'nearest')

train_generator = train_datagen.flow_from_directory('labels/train',
                                                   target_size=(150,150))


val_datagen = ImageDataGenerator(rescale =1.0/255)

val_generator = val_datagen.flow_from_directory('labels/val',
                                                   target_size=(150,150))

test_datagen = ImageDataGenerator(rescale =1.0/255)

test_generator = test_datagen.flow_from_directory('labels/test',
                                                   target_size=(150,150),
                                                 shuffle=False)
```

## 4.Building a model to train the data on. 

```python
#Callback
red_lr = ReduceLROnPlateau(monitor='val_loss',patience=5,factor=0.1,verbose=1)
```

```python
#Creating a CNN model.
model = Sequential([
    Conv2D(32, (5,5), input_shape = (150,150,3), padding='Same',activation='relu'),
    MaxPool2D(2,2),
    Conv2D(32, (3,3), padding='Same', activation='relu'),
    MaxPool2D(pool_size=(2,2), strides=(2,2)),
    Conv2D(64, (3,3), padding='Same', activation='relu'),
    MaxPool2D(pool_size=(2,2),strides=(2,2)),
    Conv2D(64, (3,3), padding='Same', activation='relu'),
    MaxPool2D(pool_size = (2,2), strides=(2,2)),
    Flatten(),
    Dense(512,activation='relu'),
    Dense(2,activation='softmax')
])
```

## 7.Evaluating our model 

```python
predictions = model.predict(test_generator)
```

```python
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)

    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j],
            horizontalalignment="center",
            color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

```python
plot_confusion_matrix(confusion_matrix(y_true=test_generator.classes, y_pred = np.argmax(predictions,axis=1)), labels)
```

```python
print(classification_report(y_true=test_generator.classes, y_pred = np.argmax(predictions,axis=1)))
```
